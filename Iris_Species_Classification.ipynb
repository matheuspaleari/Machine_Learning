```
# testando qual o melhor valor de K de 1 a 25

from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data
y = iris.target
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train = train_test_split(x,y,test_size=0.25)


```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
valores_performance = {}
K= 1

while K <= 25:
  knn = KNeighborsClassifier (n_neighbors=K)
  knn.fit(x_train,y_train)
  previsoes = knn.predict(x_test)
  acertos = metrics.accuracy_score(y_test,previsoes)
  valores_performance[K] = round(acertos,4)
  K+= 1

print(valores_performance)
```

```
import matplotlib.pyplot as plt

plt.plot(list(valores_performance.keys()),list(valores_performance.values()))
plt.xlabel('valores de K')
plt.ylabel('performance')
plt.show
```

```
### utilizando regressão logica
from numpy.random import logistic
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression() ### por padrão ele usa multi_class ="ovr"
logreg.fit(x_train,y_train)
print(logreg.predict([[5.8,2.6,4,1.2]]))
print(logreg.predict_proba([[5.8,2.6,4,1.2]]))
```

```
previsoes_logreg = logreg.predict(x_test)
acertos_logreg = metrics.accuracy_score(y_test, previsoes_logreg)
print(acertos_logreg)
```

Conseguimos validar que o KNN acaba por ser melhor que a regressão logica.
